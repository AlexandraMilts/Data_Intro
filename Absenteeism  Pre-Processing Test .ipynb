```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('Absenteeism-data.csv')
```

```python
df = data.copy()
```

```python
df = df.drop(['ID'], axis = 1)
#axis = 1 means we should drop a column, not a row
```

```python
df
```

```python
data_with_dummies_reason = pd.get_dummies(data['Reason for Absence'], drop_first=True)
```

```python
data_with_dummies_reason
```

```python
group1 = data_with_dummies_reason.loc[:,'1':'14']
group2 = data_with_dummies_reason.loc[:,'15':'17']
group3 = data_with_dummies_reason.loc[:,'18':'21']
group4 = data_with_dummies_reason.loc[:,'22':'28']
```

```python
group3
```

```python
df_without_reason = df.drop(['Reason for Absence'], axis = 1)
```

```python
df_without_reason
```

```python
reason_type_1 = group1.max(axis=1)
reason_type_2 = group2.max(axis=1)
reason_type_3 = group3.max(axis=1)
reason_type_4 = group4.max(axis=1)
```

```python
reason_type_1
```

```python
df = pd.concat([df_without_reason, reason_type_1, reason_type_2, reason_type_3, reason_type_4], axis = 1)
```

```python
df
```

```python
df.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours', 'Reason_1', 'Reason_2', 'Reason_3','Reason_4']
```

```python
df.columns = column_names
```

```python
df.head()
```

```python
df_reason = df.copy()
```

```python
df_reason['Date'] = pd.to_datetime(df_reason['Date'], format ='%d/%m/%Y')
```

```python
df_reason.head()
```

```python
df_reason['Date'][0].month
```

```python
months = []
for i in range(df_reason.shape[0]):
    months.append(df_reason['Date'][i].month)
```

```python
months
```

```python
df_reason['Month'] = months
```

```python
df_reason.head()
```

```python
def date_to_weekday(date):
    return date.weekday()
```

```python
df_reason['Day of the Week'] = df_reason['Date'].apply(date_to_weekday)
```

```python
df_reason.head()
```

```python
df_reason_dates = df_reason.drop(['Date'], axis = 1)
```

```python
df_reason_dates.head()
```

```python
df_almost = df_reason_dates.copy()
```

```python
df_almost['Education'] = df_almost['Education'].map({1:0, 2:1, 3:1, 4:1})
```
